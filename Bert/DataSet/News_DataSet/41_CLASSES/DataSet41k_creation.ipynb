```python
import pandas as pd
import numpy as np
```

```python
FILE_PATH = "/content/drive/MyDrive/GPUs/Bert/DataSet_Creation/News_DataSet/News_DataSet.csv"
SAVE_PATH = "/content/drive/MyDrive/GPUs/Bert/DataSet_Creation/News_DataSet/MySave.csv"
NUM_ARTICLES = 1000
RANDOM_SEED = 42

np.random.seed(RANDOM_SEED)
```

```python
df = pd.read_csv(FILE_PATH)
df.columns.values
```

```python
df.drop(columns=['authors', 'link', 'short_description', 'date'], inplace=True)
```

```python
df['label'] = 0
df.head()
```

```python
#DATASET 41K
className = df.category.unique()
print(className)

newsDataset = pd.DataFrame(columns=df.columns.values)
labelID = 0

for name in className:
  df_add =  df[(df.category == name)][:1000]
  df_add['label'] = labelID
  newsDataset = pd.concat([newsDataset, df_add])
  labelID += 1


newsDataset.head()
```

```python
newsDataset.to_csv("/content/drive/MyDrive/GPUs/Bert/DataSet/News_DataSet/41_CLASSES/News41k.csv", index=False)
```
